```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(0,10)
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# to set axis limit the below two lines
plt.xlim(0)
plt.ylim(0,15)
plt.title("String Title")
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
x = np.arange(0,11)
```

```python
y = 2*x
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title('Power')
axes2 = fig.add_axes([1.5,0.2,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(1, 8)
axes2.set_ylim(0, 8000)
axes2.set_xlabel('x')
axes2.set_ylabel('Y')
axes2.set_title('same')

plt.show()

# the semicolon will remove the line in the output
```

```python
fig = plt.figure(figsize=(10,3),dpi=100)
# the dpi is used to make the text more clear to view but not to set too high as the pc will take time to load
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_fig.png',bbox_inches='tight')
# the bbox_inches will display the axis
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,)
# the axes is a numpy array 
axes[0,0].plot(a,b)
axes[0,1].plot(x,y)
axes[1,0].plot(x,a)
axes[1,1].plot(y,b)
plt.tight_layout()  # prevents overlapping or close to each other
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,)
# the axes is a numpy array
axes[0, 0].plot(a, b)
axes[0,0].set_title("Title for [0, 0]")
axes[0, 1].plot(x, y)
axes[1, 0].plot(x, a)
axes[1, 1].plot(y, b)
fig.suptitle("Figure")
fig.set_figwidth(10)
fig.subplots_adjust(wspace=0.5,hspace=1)
fig.savefig("new-subplots.png",bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x vs y')
ax.plot(x,x,label='x vs x')
# ax.legend(loc='lower left')  
# ax.legend(loc=3)  
ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#458716',label='x vs x',lw=5,ls='--')
lines = ax.plot(x,x+1,color='#f56716',label='x vs y',linewidth=6)
lines[0].set_dashes([5,2,10,2])
# 5 long line and 2 spaces and 10 long line and 2 spaces of custom styling
```

```python
fig = plt.figure(figsize=(14,7))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, color='#458716', label='x vs x', lw=5, ls='--')
lines = ax.plot(x, x+1, color='#f56716', label='x vs x+1', linewidth=1,marker='o',ls='--')
lines = ax.plot(x, x+2, color='red', label='x vs x+2',
                linewidth=1, marker='o', ls='--',markersize=20,markerfacecolor='blue',markeredgewidth=6,markeredgecolor='orange')

# markers 
```
